# HR Employee Attrition - Exploratory Data Analysis (EDA)

This notebook provides a detailed walkthrough of the exploratory data analysis for the IBM HR Employee Attrition dataset, documenting findings, target imbalance, feature distributions, and business-focused correlations.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

# Set style for figures
sns.set_theme(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)
plt.rcParams["font.size"] = 12
```

## 1. Load and Inspect Dataset

```python
data_path = "../data/WA_Fn-UseC_-HR-Employee-Attrition.csv"
df = pd.read_csv(data_path)
print(f"Dataset Shape: {df.shape}")
df.head()
```

## 2. Data Validation & Summary Stats

```python
print("Missing values:", df.isnull().sum().sum())
print("Duplicate rows:", df.duplicated().sum())

# Check for constant columns
constant_cols = [col for col in df.columns if df[col].nunique() <= 1]
print("Constant/Redundant columns to drop:", constant_cols)
```

## 3. Class Imbalance (Attrition Target)

```python
attrition_counts = df["Attrition"].value_counts()
attrition_pct = df["Attrition"].value_counts(normalize=True) * 100

fig, ax = plt.subplots(figsize=(6, 4))
sns.barplot(x=attrition_counts.index, y=attrition_counts.values, palette=["#3b82f6", "#ef4444"], ax=ax)
for i, val in enumerate(attrition_counts.values):
    ax.text(i, val + 15, f"{val} ({attrition_pct.iloc[i]:.1f}%)", ha='center', fontweight='bold')
ax.set_title("Target Class Distribution (Attrition)")
ax.set_ylabel("Count")
plt.show()
```

## 4. Key Business Correlations

### A. Attrition by Overtime Working

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x="OverTime", hue="Attrition", palette=["#3b82f6", "#ef4444"])
plt.title("Overtime vs Attrition")
plt.show()
```

### B. Attrition by Monthly Income

```python
plt.figure(figsize=(7, 4.5))
sns.boxplot(data=df, x="Attrition", y="MonthlyIncome", palette=["#3b82f6", "#ef4444"])
plt.title("Monthly Income vs Attrition")
plt.show()
```

### C. Attrition by Work-Life Balance and Job Satisfaction

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

sns.barplot(data=df, x="WorkLifeBalance", y=df["Attrition"].map({"Yes":1, "No":0}), ax=axes[0], color="#ef4444", errorbar=None)
axes[0].set_title("Attrition Rate vs Work Life Balance")
axes[0].set_ylabel("Attrition Rate")
axes[0].yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x:.0%}'))

sns.barplot(data=df, x="JobSatisfaction", y=df["Attrition"].map({"Yes":1, "No":0}), ax=axes[1], color="#ef4444", errorbar=None)
axes[1].set_title("Attrition Rate vs Job Satisfaction")
axes[1].set_ylabel("Attrition Rate")
axes[1].yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x:.0%}'))

plt.tight_layout()
plt.show()
```

## 5. Numerical Feature Correlation Heatmap

```python
# Filter numerical features excluding constants and IDs
num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
clean_num_cols = [c for c in num_cols if c not in ["EmployeeCount", "StandardHours", "EmployeeNumber"]]

corr_matrix = df[clean_num_cols].corr()

plt.figure(figsize=(14, 12))
sns.heatmap(corr_matrix, annot=False, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap of Numerical Features")
plt.show()
```
